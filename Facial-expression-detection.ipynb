### Splitting the data

```python
# Split the data based on Usage
train_df = df[df['Usage'] == 'Training']
val_df = df[df['Usage'] == 'PublicTest']
test_df = df[df['Usage'] == 'PrivateTest']
```

```python
print('Shape of train_df : ',train_df.shape)
print('Shape of val_df : ',val_df.shape )
print('shape fo test_df : ',test_df.shape)
```

## Cnvert pixels to images

```python
import numpy as np

# Function to process pixel data
def process_pixels(pixel_data):
    return np.array([np.fromstring(p, dtype=np.float32, sep=' ').reshape(48, 48) for p in pixel_data])

# Apply to each dataset
X_train = process_pixels(train_df['pixels'])
X_val = process_pixels(val_df['pixels'])
X_test = process_pixels(test_df['pixels'])

# Normalize pixel values to [0, 1]
X_train = X_train / 255.0
X_val = X_val / 255.0
X_test = X_test / 255.0
```

### Extract the labels

```python
y_train = train_df['emotion'].values
y_val = val_df['emotion'].values
y_test = test_df['emotion'].values
```

### Flatten the images for pca

```python
# Flatten the images for PCA
X_train_flat = X_train.reshape(X_train.shape[0], -1)  # Shape: (n_samples, 2304)
X_val_flat = X_val.reshape(X_val.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

```

## Plot some images

```python
# Create a mapping from label to emotion
emotion_labels = {
    0: 'Anger',
    1: 'Disgust',
    2: 'Fear',
    3: 'Happy',
    4: 'Sad',
    5: 'Surprise',
    6: 'Neutral'
}

# Plot the images with the mapped emotion labels
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(X_train[i].squeeze(), cmap='gray')
    plt.title(f"{emotion_labels[y_train[i]]}")  # Using the mapping
    plt.axis('off')
plt.tight_layout()
plt.show()
```

### Applying the PCA technique for feature extraction and also for dimensionality reduction

```python
from sklearn.decomposition import PCA

n_components = 80
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_flat)
X_val_pca = pca.transform(X_val_flat)
X_test_pca = pca.transform(X_test_flat)

print(f"Explained variance with {n_components} components: {np.sum(pca.explained_variance_ratio_):.2f}")

```

### Applying the suppoert vector classification

```python
# Train and evaluate SVC
from sklearn.svm import SVC
svc = SVC(random_state=42)
svc.fit(X_train_pca, y_train)
```

```python

y_val_pred = svc.predict(X_val_pca)
```

### Accuracy_score

```python
# Compute accuracy
from sklearn.metrics import accuracy_score
val_accuracy = accuracy_score(y_val, y_val_pred)
print(f"Validation Accuracy: {val_accuracy:.2f}")
```

### Classification report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
# Compute confusion matrix
conf_matrix = confusion_matrix(y_val, y_val_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='coolwarm', cbar=False)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix - Validation Set")
plt.show()
```
